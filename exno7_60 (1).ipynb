```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Step 1: Load built-in MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Step 2: Preprocess data
# Normalize pixel values to [0,1]
x_train = x_train / 255.0
x_test = x_test / 255.0

# Add channel dimension (needed for CNNs)
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# Step 3: Build CNN model
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')  # 10 classes (digits 0–9)
])

# Step 4: Compile model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Step 5: Train model
model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))

# Step 6: Evaluate model
loss, acc = model.evaluate(x_test, y_test)
print(f"Test accuracy: {acc:.2f}")

# Step 7: Predict on a single test image
import numpy as np
sample = x_test[0].reshape(1,28,28,1)  # take first test image
prediction = model.predict(sample)
predicted_class = np.argmax(prediction)
print("Predicted digit:", predicted_class)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np

# Step 1: Load Fashion-MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

# Step 2: Preprocess data
x_train = x_train / 255.0
x_test = x_test / 255.0

# Add channel dimension (needed for CNNs)
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# Step 3: Build CNN model
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')  # 10 clothing categories
])

# Step 4: Compile model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Step 5: Train model
model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))

# Step 6: Evaluate model
loss, acc = model.evaluate(x_test, y_test)
print(f"Test accuracy: {acc:.2f}")

# Step 7: Predict on a single test image
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
               "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]

sample = x_test[0].reshape(1,28,28,1)  # take first test image
prediction = model.predict(sample)
predicted_class = class_names[np.argmax(prediction)]
print("Predicted clothing item:", predicted_class)
import matplotlib.pyplot as plt

# Train model and store history
history = model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))

# Plot accuracy
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Plot loss
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Load MNIST
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train = x_train.reshape(-1, 784) / 255.0
x_test = x_test.reshape(-1, 784) / 255.0

# Binary classification: digit 0 vs not 0
y_train = (y_train == 0).astype(np.float32)
y_test = (y_test == 0).astype(np.float32)

# Logistic regression model (no bias)
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, activation='sigmoid', use_bias=False, input_shape=(784,))
])

# Try different optimizers
optimizers = {
    "SGD": tf.keras.optimizers.SGD(),
    "Nesterov": tf.keras.optimizers.SGD(momentum=0.9, nesterov=True),
    "Adagrad": tf.keras.optimizers.Adagrad(),
    "RMSProp": tf.keras.optimizers.RMSprop(),
    "Adam": tf.keras.optimizers.Adam()
}

histories = {}
for name, opt in optimizers.items():
    model_copy = tf.keras.models.clone_model(model)
    model_copy.compile(optimizer=opt,
                       loss='binary_crossentropy',
                       metrics=['accuracy'])
    history = model_copy.fit(x_train, y_train,
                             validation_data=(x_test, y_test),
                             epochs=5, batch_size=128, verbose=0)
    histories[name] = history

# Plot loss vs. iteration
for name, history in histories.items():
    plt.plot(history.history['loss'], label=f'{name} Train')
    plt.plot(history.history['val_loss'], label=f'{name} Val')
plt.title("Loss vs Epochs (MNIST Logistic Regression)")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# -----------------------------
# Step 1: Load MNIST (binary classification: 0 vs not 0)
# -----------------------------
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train = x_train.reshape(-1, 784).astype("float32") / 255.0
x_test = x_test.reshape(-1, 784).astype("float32") / 255.0
y_train = (y_train == 0).astype("float32")
y_test = (y_test == 0).astype("float32")

# -----------------------------
# Step 2: Define logistic regression model
# -----------------------------
def build_model():
    return tf.keras.Sequential([
        tf.keras.layers.Dense(1, activation="sigmoid", use_bias=False, input_shape=(784,))
    ])

# -----------------------------
# Step 3: Train with Adam optimizer
# -----------------------------
model = build_model()
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              loss="binary_crossentropy",
              metrics=["accuracy"])
history = model.fit(x_train, y_train, epochs=3, batch_size=128, verbose=0)

# -----------------------------
# Step 4: Extract weights
# -----------------------------
weights = model.layers[0].get_weights()[0].flatten()

# -----------------------------
# Step 5: Reduce dimensionality (PCA to 2D)
# -----------------------------
from sklearn.decomposition import PCA
weights = model.layers[0].get_weights()[0].flatten().reshape(1, -1)  # shape (1, 784)
pca = PCA(n_components=2)
weights_2d = pca.fit_transform(weights)  # now you have (1,2)

# -----------------------------
# Step 6: Compute loss surface in reduced space
# -----------------------------
grid_x, grid_y = np.meshgrid(np.linspace(-2,2,50), np.linspace(-2,2,50))
loss_surface = np.zeros_like(grid_x)

for i in range(grid_x.shape[0]):
    for j in range(grid_x.shape[1]):
        # reconstruct approximate weights from PCA space
        w_proj = pca.inverse_transform([[grid_x[i,j]], [grid_y[i,j]]]).flatten()
        logits = tf.matmul(x_test, tf.constant(w_proj.reshape(-1,1), dtype=tf.float32))
        preds = tf.sigmoid(logits)
        loss = tf.reduce_mean(tf.keras.losses.binary_crossentropy(y_test, preds))
        loss_surface[i,j] = loss.numpy()

# -----------------------------
# Step 7: Plot 3D loss surface
# -----------------------------
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(grid_x, grid_y, loss_surface, cmap="viridis", alpha=0.8)
ax.set_title("3D Loss Surface (MNIST Logistic Regression)")
ax.set_xlabel("PCA Dimension 1")
ax.set_ylabel("PCA Dimension 2")
ax.set_zlabel("Loss")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import tensorflow as tf
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D

# ============================================================
# PART 1: Analytical Objective Function
# ============================================================

def f(w1, w2):
    return (w1**2 + w2 - 11)**2 + (w1 + w2**2 - 7)**2

def grad(w1, w2):
    df_dw1 = 4*w1*(w1**2 + w2 - 11) + 2*(w1 + w2**2 - 7)
    df_dw2 = 2*(w1**2 + w2 - 11) + 4*w2*(w1 + w2**2 - 7)
    return np.array([df_dw1, df_dw2])

# Optimizer example: Gradient Descent
def gradient_descent(init, lr=0.01, steps=50):
    w = np.array(init, dtype=float)
    trajectory = [w.copy()]
    for _ in range(steps):
        g = grad(w[0], w[1])
        w -= lr * g
        trajectory.append(w.copy())
    return np.array(trajectory)

trajectory = gradient_descent([0.0, 0.0])

# 2D contour animation
w1 = np.linspace(-5, 5, 200)
w2 = np.linspace(-5, 5, 200)
W1, W2 = np.meshgrid(w1, w2)
Z = f(W1, W2)

fig, ax = plt.subplots()
ax.contour(W1, W2, Z, levels=50)
line, = ax.plot([], [], 'ro-')

def update(i):
    line.set_data(trajectory[:i,0], trajectory[:i,1])
    return line,

ani = FuncAnimation(fig, update, frames=len(trajectory), interval=200)
plt.show()

# 3D surface plot
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(W1, W2, Z, cmap="viridis", alpha=0.8)
ax.plot(trajectory[:,0], trajectory[:,1], [f(w[0],w[1]) for w in trajectory], 'r.-')
ax.set_title("3D Loss Surface with Optimizer Trajectory")
plt.show()

# ============================================================
# PART 2: MNIST Logistic Regression
# ============================================================

# Load MNIST
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train = x_train.reshape(-1, 784).astype("float32") / 255.0
x_test = x_test.reshape(-1, 784).astype("float32") / 255.0
y_train = (y_train == 0).astype("float32")
y_test = (y_test == 0).astype("float32")

# Logistic regression model
def build_model(opt):
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(1, activation="sigmoid", use_bias=False, input_shape=(784,))
    ])
    model.compile(optimizer=opt, loss="binary_crossentropy", metrics=["accuracy"])
    return model

optimizers = {
    "SGD": tf.keras.optimizers.SGD(0.01),
    "Nesterov": tf.keras.optimizers.SGD(0.01, momentum=0.9, nesterov=True),
    "Adagrad": tf.keras.optimizers.Adagrad(0.01),
    "RMSProp": tf.keras.optimizers.RMSprop(0.001),
    "Adam": tf.keras.optimizers.Adam(0.001)
}

histories = {}
weight_histories = {}

for name, opt in optimizers.items():
    model = build_model(opt)
    weight_history = []
    history = model.fit(x_train, y_train, epochs=5, batch_size=128,
                        validation_data=(x_test, y_test), verbose=0,
                        callbacks=[tf.keras.callbacks.LambdaCallback(
                            on_epoch_end=lambda epoch, logs: weight_history.append(
                                model.layers[0].get_weights()[0].flatten()))])
    histories[name] = history
    weight_histories[name] = np.array(weight_history)

# Loss vs iteration plot
plt.figure(figsize=(10,6))
for name, history in histories.items():
    plt.plot(history.history["loss"], label=f"{name} Train")
    plt.plot(history.history["val_loss"], linestyle="--", label=f"{name} Val")
plt.title("Loss vs Epochs (MNIST Logistic Regression)")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()

# 3D PCA trajectory plot for Adam
pca = PCA(n_components=3)
weights_3d = pca.fit_transform(weight_histories["Adam"])

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection="3d")
ax.plot(weights_3d[:,0], weights_3d[:,1], weights_3d[:,2], marker='o')
ax.set_title("Adam Optimizer Trajectory in PCA-Reduced 3D Space (MNIST)")
ax.set_xlabel("PCA Dim 1")
ax.set_ylabel("PCA Dim 2")
ax.set_zlabel("PCA Dim 3")
plt.show()
```
