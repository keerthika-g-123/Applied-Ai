```python
import math
def sigmoid(z):
    return 1 / (1 + math.exp(-z))
# Inputs
x1, x2, x3, x4 = 1, 1, 0, 1
# Initial Weights
W15, W16 = 0.3, 0.1
W25, W26 = -0.2, 0.4
W35, W36 = 0.2, -0.3
W45, W46 = 0.1, 0.4
W57, W67 = -0.3, 0.2
# Biases
B5, B6, B7 = 0.2, 0.1, -0.3
# Learning rate
lr = 0.8
# Target output
target = 1
# Training loop
epoch = 0
while True:
    epoch += 1
    # ---- Forward Propagation ----
    net5 = x1*W15 + x2*W25 + x3*W35 + x4*W45 + B5
    x5 = sigmoid(net5)

    net6 = x1*W16 + x2*W26 + x3*W36 + x4*W46 + B6
    x6 = sigmoid(net6)

    net7 = x5*W57 + x6*W67 + B7
    x7 = sigmoid(net7)
    # Error
    error = target - x7
    # Stop condition
    if abs(error) < 0.01:
        print(f"Training converged at epoch {epoch}")
        print("Final Output:", x7)
        break
    # ---- Backward Propagation ----
    delta7 = error * x7 * (1-x7)
    delta5 = x5 * (1-x5) * (delta7 * W57)
    delta6 = x6 * (1-x6) * (delta7 * W67)
    # Update weights and biases
    W57 += lr * delta7 * x5
    W67 += lr * delta7 * x6
    B7  += lr * delta7

    W15 += lr * delta5 * x1
    W25 += lr * delta5 * x2
    W35 += lr * delta5 * x3
    W45 += lr * delta5 * x4
    B5  += lr * delta5

    W16 += lr * delta6 * x1
    W26 += lr * delta6 * x2
    W36 += lr * delta6 * x3
    W46 += lr * delta6 * x4
    B6  += lr * delta6
# Print final weights
print("Updated Weights and Biases:")
print("W57:", W57, "W67:", W67, "B7:", B7)
print("W15:", W15, "W25:", W25, "W35:", W35, "W45:", W45, "B5:", B5)
print("W16:", W16, "W26:", W26, "W36:", W36, "W46:", W46, "B6:", B6)
```
