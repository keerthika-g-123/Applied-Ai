```python
import numpy as np
def step_function(net):
    return 1 if net >= 0 else 0 
def train_perceptron(X, y, w, b, lr, gate_name):
    epoch = 0   
    print(f"\nTraining {gate_name} Gate")
    while True:
        error_count = 0
        print(f"\nEpoch {epoch + 1}")
        for i in range(len(X)):
            x = X[i]
            target = y[i]
            net = np.dot(w, x) -b
            output = step_function(net)
            error = target - output
            
            print(f"Input: {x}, Net: {net:.2f}, Output: {output}, Target: {target}")
            if error != 0:
                w = w + lr * error * x
                b = b - lr * error
                error_count += 1
                print(f"  Updated Weights: {w}, Updated Bias: {b:.2f}")

        epoch += 1       
        if error_count == 0:
            break
        print(f"Final Weights:{w} Final Bias:{b}")

    return w, b
X_or = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
y_or = np.array([0, 1, 1, 1])
w_or = np.array([0.2, 0.2])
b_or = -0.4
learning_rate = 0.1
final_w_or, final_b_or = train_perceptron(
    X_or, y_or, w_or, b_or, learning_rate, "OR"
)
y_and=np.array([0,0,0,1])
w_and=np.array([0.2,0.2])
b_and=-0.3
X_and = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
y_and = np.array([0, 0, 0, 1])
w_and = np.array([0.2, 0.2])
b_and = -0.3
final_w_and, final_b_and = train_perceptron(
    X_and, y_and, w_and, b_and, learning_rate, "AND"
)
```
