```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_auc_score,
    average_precision_score,
    roc_curve,
    precision_recall_curve
)

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping


# -----------------------------
# 1. LOAD DATA
# -----------------------------
df = pd.read_csv(r"c:\Users\muddu\Downloads\archive.zip")

X = df.drop("Class", axis=1)
y = df["Class"]

# -----------------------------
# 2. SCALE FEATURES
# -----------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# -----------------------------
# 3. CLASS WEIGHT
# -----------------------------
neg, pos = np.bincount(y)
class_weight = {0: 1.0, 1: neg / pos}

# -----------------------------
# 4. MODEL DEFINITION (3 hidden layers)
# -----------------------------
def build_dnn(input_dim, learning_rate=0.001):
    model = Sequential([
        Dense(128, activation='relu', input_dim=input_dim),
        BatchNormalization(),
        Dropout(0.4),

        Dense(64, activation='relu'),
        BatchNormalization(),
        Dropout(0.3),

        Dense(32, activation='relu'),
        BatchNormalization(),
        Dropout(0.2),

        Dense(1, activation='sigmoid')
    ])

    model.compile(
        optimizer=Adam(learning_rate=learning_rate),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model


# -----------------------------
# 5. EVALUATION FUNCTIONS
# -----------------------------
def plot_confusion_matrix(cm):
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
def evaluate_model(model, X_test, y_test, title=""):
    y_prob = model.predict(X_test).ravel()
    y_pred = (y_prob > 0.5).astype(int)

    print("\n" + title)
    print(classification_report(y_test, y_pred))

    cm = confusion_matrix(y_test, y_pred)
    plot_confusion_matrix(cm)

    roc_auc = roc_auc_score(y_test, y_prob)
    pr_auc = average_precision_score(y_test, y_prob)

    print(f"ROC-AUC: {roc_auc:.4f}")
    print(f"PR-AUC: {pr_auc:.4f}")

    fpr, tpr, _ = roc_curve(y_test, y_prob)
    precision, recall, _ = precision_recall_curve(y_test, y_prob)

    plt.plot(fpr, tpr)
    plt.title("ROC Curve")
    plt.show()

    plt.plot(recall, precision)
    plt.title("Precision-Recall Curve")
    plt.show()

    return roc_auc, pr_auc


# -----------------------------
# 6. EXPERIMENT 1: 80-20 split
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42
)

model = build_dnn(X_train.shape[1])
history = model.fit(
    X_train, y_train,
    epochs=20,
    batch_size=2048,
    class_weight=class_weight,
    verbose=0
)

plt.plot(history.history['loss'])
plt.title("Experiment 1 - Training Loss Curve")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()

exp1_roc, exp1_pr = evaluate_model(model, X_test, y_test, "Experiment 1: 80–20 Split")


# -----------------------------
# 7. EXPERIMENT 2: 70-30 split
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, stratify=y, random_state=42
)

model = build_dnn(X_train.shape[1])
history = model.fit(
    X_train, y_train,
    epochs=20,
    batch_size=2048,
    class_weight=class_weight,
    verbose=0
)

plt.plot(history.history['loss'])
plt.title("Experiment 2 - Training Loss Curve")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()

exp2_roc, exp2_pr = evaluate_model(model, X_test, y_test, "Experiment 2: 70–30 Split")


# -----------------------------
# 8. EXPERIMENT 3: 60-20-20 split
# -----------------------------
X_temp, X_test, y_temp, y_test = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42
)

X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.25, stratify=y_temp, random_state=42
)

model = build_dnn(X_train.shape[1])
es = EarlyStopping(patience=3, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    epochs=20,
    validation_data=(X_val, y_val),
    batch_size=2048,
    class_weight=class_weight,
    callbacks=[es],
    verbose=0
)

plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Val')
plt.title("Experiment 3 - Loss Curves")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()

plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Val')
plt.title("Experiment 3 - Accuracy Curves")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

exp3_roc, exp3_pr = evaluate_model(model, X_test, y_test, "Experiment 3: 60–20–20 Split")


# -----------------------------
# 9. EXPERIMENT 4: 70-15-15 split
# -----------------------------
X_temp, X_test, y_temp, y_test = train_test_split(
    X_scaled, y, test_size=0.15, stratify=y, random_state=42
)

X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.176, stratify=y_temp, random_state=42
)

model = build_dnn(X_train.shape[1])
es = EarlyStopping(patience=3, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    epochs=20,
    validation_data=(X_val, y_val),
    batch_size=2048,
    class_weight=class_weight,
    callbacks=[es],
    verbose=0
)

plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Val')
plt.title("Experiment 4 - Loss Curves")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()

plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Val')
plt.title("Experiment 4 - Accuracy Curves")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

exp4_roc, exp4_pr = evaluate_model(model, X_test, y_test, "Experiment 4: 70–15–15 Split")


# -----------------------------
# 10. BEST MODEL SELECTION (based on PR-AUC)
# -----------------------------
pr_scores = [exp1_pr, exp2_pr, exp3_pr, exp4_pr]
best_idx = np.argmax(pr_scores)
print("\nBest Experiment:", best_idx + 1)
print("Best PR-AUC:", pr_scores[best_idx])


# -----------------------------
# 11. 5-FOLD STRATIFIED CROSS VALIDATION
# -----------------------------
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

roc_scores = []
pr_scores = []

for train_idx, test_idx in skf.split(X_scaled, y):
    X_train, X_test = X_scaled[train_idx], X_scaled[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    model = build_dnn(X_train.shape[1])
    model.fit(
        X_train, y_train,
        epochs=20,
        batch_size=2048,
        class_weight=class_weight,
        verbose=0
    )

    y_prob = model.predict(X_test).ravel()
    roc_scores.append(roc_auc_score(y_test, y_prob))
    pr_scores.append(average_precision_score(y_test, y_prob))

print("\n5-Fold Cross Validation Results")
print(f"ROC-AUC: {np.mean(roc_scores):.4f} ± {np.std(roc_scores):.4f}")
print(f"PR-AUC: {np.mean(pr_scores):.4f} ± {np.std(pr_scores):.4f}")
```
